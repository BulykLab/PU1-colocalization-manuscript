```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
# bqtl[0] is the list of significant bQTL
bqtl = pd.read_table('PU1.nominal.lead.centered.noSV.K10.txt', delimiter=' ', header=None)
```

```python
bqtl_dict = {}
for i in range(bqtl.shape[0]):
    peakname = bqtl[0][i]
    varpos = bqtl[9][i]
    pval = bqtl[11][i]
    rsid = bqtl[7][i]
    bqtl_dict[peakname] = [varpos,pval,rsid]
```

```python
motifdisrupting = pd.read_table('PU1.pwm_gkmsvm.nominal.txt', delimiter='\t', header=None)
```

```python
all_bqtl = pd.read_table('PU1.nominal.all.variants.txt', delimiter=' ', header=None)
ld = pd.read_table('PU1.leadsnps.r2_0.8.ld.txt', delimiter='\t', header=0)
```

```python
ld_dict = {}
for i in range(ld.shape[0]):
    if ld['SNP_A'][i] in ld_dict:
        ld_dict[ld['SNP_A'][i]].append(ld['SNP_B'][i])
    else:            
        ld_dict[ld['SNP_A'][i]] = [ld['SNP_B'][i]]
```

```python
closest_bqtl_dict = {}
#print(all_bqtl.head())
for i in range(all_bqtl.shape[0]):
    peakname = all_bqtl[0][i]
    varpos = all_bqtl[9][i]
    pval = all_bqtl[11][i]
    rsid = all_bqtl[7][i]
    dist = all_bqtl[6][i]
    if peakname in bqtl_dict:
        if rsid in ld_dict[bqtl_dict[peakname][2]]:
            if peakname not in closest_bqtl_dict:
                closest_bqtl_dict[peakname] = [varpos,pval,rsid,dist]
            else:
                if abs(dist) < abs(closest_bqtl_dict[peakname][3]):
                    closest_bqtl_dict[peakname] = [varpos,pval,rsid,dist]
```

```python
mask = []
vals = []
count = 0
pu1motif_dict = {}
for i in range(motifdisrupting.shape[0]):
    peakname = motifdisrupting[3][i]
    pval = motifdisrupting[17][i]
    ref = motifdisrupting[12][i]
    alt = motifdisrupting[13][i]
    rsid = motifdisrupting[4][i]
    
    test = False
    if peakname in list(bqtl[0]):
        if pval < 5e-2 and min(ref, alt) > 0:
            if rsid in ld_dict[bqtl_dict[peakname][2]]:
                dist = motifdisrupting[5][i] - motifdisrupting[1][i] - 100 - 1
                test = True
                if peakname not in pu1motif_dict:
                    pu1motif_dict[peakname] = dist
                elif abs(dist) < abs(pu1motif_dict[peakname]):
                    pu1motif_dict[peakname] = dist
    mask.append(test)

mask = pd.Series(mask)
```

```python
vals = []
for peak in pu1motif_dict:
    vals.append(pu1motif_dict[peak])
```

```python
plt.figure(figsize = (15, 10)) 

# We can set the number of bins with the `bins` kwarg
counts = plt.hist(vals, bins=np.array(range(-100,100)))#, rwidth=0.55, color='black')
#axs[1].hist(blood_total, bins=100)

#plt.axvline(x = 94, color = 'r')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.xlabel('# of variants tagging (r2>0.8) blood cell trait associations', fontsize = 16)
plt.show()
```

```python
counts = [0,0,0,0,0]
for peak in closest_bqtl_dict:
    if peak in pu1motif_dict:
        if abs(pu1motif_dict[peak]) <= 25:
            counts[0] += 1
        else:
            counts[1] += 1
    else:
        if abs(closest_bqtl_dict[peak][3]) <= 25:
            counts[2] += 1
        elif abs(closest_bqtl_dict[peak][3]) <= 100:
            counts[3] += 1
        else:
            counts[4] += 1
```

```python
mask = []
vals = []
count = 0
for i in range(motifdisrupting.shape[0]):
    if motifdisrupting[3][i] in list(bqtl[0]):
        if motifdisrupting[17][i] < 5e-2 and min(motifdisrupting[12][i], motifdisrupting[13][i]) > 0:
            #if motifdisrupting[17][i] > bqtl_dict[motifdisrupting[3][i]][1]:
            if (motifdisrupting[5][i] == bqtl_dict[motifdisrupting[3][i]][0]) or (motifdisrupting[17][i] < bqtl_dict[motifdisrupting[3][i]][1] * 15):
                count += 1
                #print([i])
                vals.append(motifdisrupting[5][i] - motifdisrupting[1][i] -100 - 1)
                mask.append(True)
                #if motifdisrupting[5][i] != bqtl_dict[motifdisrupting[3][i]][0]:
                #    print(motifdisrupting[4][i], bqtl_dict[motifdisrupting[3][i]][2])
                #    print(motifdisrupting[17][i], bqtl_dict[motifdisrupting[3][i]][1])
            else:
                print(motifdisrupting[4][i], bqtl_dict[motifdisrupting[3][i]][2])
                print(motifdisrupting[5][i], bqtl_dict[motifdisrupting[3][i]][0])
                print(motifdisrupting[17][i], bqtl_dict[motifdisrupting[3][i]][1])
                mask.append(False)
        else:
            mask.append(False)
        #mask.append(np.repeat(True,motifdisrupting.shape[1]))
    else:
        mask.append(False)
        #mask.append(np.repeat(False,motifdisrupting.shape[1]))
mask = pd.Series(mask)
# print(sum(mask))        

print(count)
motifdisrupting[mask]
```

```python
mask = []
vals = []
for i in range(motifdisrupting.shape[0]):
    if motifdisrupting[3][i] in list(bqtl[0]):
        if motifdisrupting[17][i] < 5e-2 and min(motifdisrupting[12][i], motifdisrupting[13][i]) > 0:
            vals.append(motifdisrupting[5][i] - motifdisrupting[1][i] -100 - 1)
            mask.append(True)
        else:
            mask.append(False)
        #mask.append(np.repeat(True,motifdisrupting.shape[1]))
    else:
        mask.append(False)
        #mask.append(np.repeat(False,motifdisrupting.shape[1]))
mask = pd.Series(mask)
# print(sum(mask))        

motifdisrupting[mask]
```

```python
# Sliding window of 3bp
sliding = np.copy(counts[0])

sliding[0] = (counts[0][0] + counts[0][1]) / 2
sliding[-1] = (counts[0][-1] + counts[0][-2]) / 2
sliding[1] = (counts[0][0] + counts[0][1] + counts[0][2]) / 2
for i in range(1,len(sliding)-1):
    sliding[i] = (counts[0][i-1] + counts[0][i] + counts[0][i+1]) / 3
plt.figure(figsize = (15, 10))
plt.xlabel('bp position from the center of the peak', fontsize = 16)
plt.ylabel('# of motif-disrupting variants (average in sliding 5bp window)', fontsize = 16)
plt.xticks(fontsize=14)
plt.yticks(range(0,19,3),fontsize=14)
plt.plot(list(range(-99,100)),sliding, color='black')
```
